```python
import pandas as pd

bmc_df = pd.read_csv("../csv/bmc-clean.csv", index_col=0)
bmc_df.reset_index(drop=True, inplace=True)
bmc_df
```

```python
bmc_df = bmc_df.dropna()
bmc_df
```

```python
bmc_df['preprocessed_content'].isnull().sum()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
import numpy as np

docs = bmc_df['preprocessed_content']
vectorizer = CountVectorizer()
X = vectorizer.fit_transform(docs)
X
```

```python
vocab = set(word for word in docs)
vocab
```

```python
n_terms = len(vocab)
n_terms
```

```python
from gsdmm import MovieGroupProcess

# Train the GSDMM
mgp = MovieGroupProcess(K=10, alpha=0.1, beta=1, n_iters=3)
mgp.fit(docs, n_terms)
```

```python
def top_words(cluster_word_distribution, top_cluster, values):
    for cluster in top_cluster:
        sort_dicts =sorted(mgp.cluster_word_distribution[cluster].items(), key=lambda k: k[1], reverse=True)[:values]
        print("\nCluster %s : %s"%(cluster,sort_dicts))
```

```python
doc_count = np.array(mgp.cluster_doc_count)
print('Number of documents per topic :', doc_count)

# topics sorted by the number of document they are allocated to
top_index = doc_count.argsort()[-10:][::-1]
print('\nMost important clusters (by number of docs inside):', top_index)
# show the top 5 words in term frequency for each cluster 
top_words(mgp.cluster_word_distribution, top_index, 10)
```

```python
top_index = doc_count.argsort()[-15:][::-1]
print('Most important clusters (by number of docs inside):', top_index)
```

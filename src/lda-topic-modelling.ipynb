```python
import pandas as pd

bmc_df = pd.read_csv("../csv/bmc-clean.csv", index_col=0)
bmc_df.reset_index(drop=True, inplace=True)
bmc_df
```

```python
bmc_df = bmc_df.dropna()
bmc_df
```

```python
# Convert Date to DateTime format
bmc_df['Date'] = pd.to_datetime(bmc_df['Date'])
bmc_df
```

```python
# group by date
grouped = bmc_df.groupby('Date')
grouped
```

```python
from gensim import corpora, models
from gensim.models import CoherenceModel
from collections import defaultdict
import numpy as np

# Get the optimal number of topics for each date
def get_optimal_num_topics(group):
    texts = [doc.split() for doc in group['preprocessed_content']]
    dictionary = corpora.Dictionary(texts)
    corpus = [dictionary.doc2bow(text) for text in texts]
    
    coherence_scores = defaultdict(list)
    
    for num_topics in range(2, 10):
        lda_model = models.LdaModel(corpus, num_topics=num_topics, id2word=dictionary, passes=10)
        coherence_model_lda = CoherenceModel(model=lda_model, texts=texts, dictionary=dictionary, coherence='c_v')
        coherence_score = coherence_model_lda.get_coherence()
        coherence_scores[num_topics].append(coherence_score)

    optimal_num_topics = max(coherence_scores, key=lambda k: np.mean(coherence_scores[k]))
    return optimal_num_topics

topics_df = pd.DataFrame(columns=['Date', 'Topics'])
```

```python
# Create dictionary and corpus for each group
results = []

# loop through the grouped data and create LDA model for each date
for date, group in grouped:
    print(f"Starting with Date: {date}")

    # get the optimal number of topics for this date
    optimal_num_topics = get_optimal_num_topics(group)
    print(f"Optimal number of topics: {optimal_num_topics}")

    # create the dictionary and corpus
    texts = [doc.split() for doc in group['preprocessed_content']]

    dictionary = corpora.Dictionary(texts)
    corpus = [dictionary.doc2bow(text) for text in texts]

    # create the LDA model
    lda_model = models.LdaModel(corpus, num_topics=optimal_num_topics, id2word=dictionary, passes=10)

    # Print all topics
    for topic in lda_model.print_topics():
        print(topic)

    # calculate the coherence score for the model
    coherence_model_lda = CoherenceModel(model=lda_model, texts=texts, dictionary=dictionary, coherence='c_v')
    coherence_score = coherence_model_lda.get_coherence()
    print(f"Coherence Score: {coherence_score}")

    # Get topics and weights for each document
    doc_topics = [lda_model.get_document_topics(doc) for doc in corpus]
    topic_weights = [[(topic[0], topic[1]) for topic in topics] for topics in doc_topics]

    print("\n")

    # Get highest scored topic for each document
    highest_topics = []
    for doc_topics in topic_weights:
        highest_topic = max(doc_topics, key=lambda x: x[1])[0]
        highest_topics.append(highest_topic)
    
    # Add results to list
    results.append({'Date': date, 'Topics': highest_topics})
```
